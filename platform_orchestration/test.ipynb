```python
import importlib
import pkgutil

from crm_agent.db.base import Base

MODEL_PACKAGES = [
    "crm_agent.db.models",
    "layer_2_account_agent.db.model",
    "layer_2_escalation_agent.db.model",
    "layer_2_proactive_agent.database.model",
    "layer_2_refund.database.model",
    "layer_2_triage.database.model",
    "layer_3.database.models",
    "platform_orchestration.models",
]

print("\n========== Importing Models ==========\n")

for package_name in MODEL_PACKAGES:
    try:
        package = importlib.import_module(package_name)
        print(f"[PACKAGE] {package_name}")

        if hasattr(package, "__path__"):
            for _, module_name, _ in pkgutil.walk_packages(
                package.__path__,
                package.__name__ + "."
            ):
                importlib.import_module(module_name)
                print(f"   ✓ {module_name}")

    except Exception as e:
        print(f"❌ Failed to import {package_name}")
        print(e)

print("\n========== Registered SQLAlchemy Tables ==========\n")

for table_name in sorted(Base.metadata.tables.keys()):
    print(f"✓ {table_name}")

print(f"\nTotal Tables Registered: {len(Base.metadata.tables)}")
```

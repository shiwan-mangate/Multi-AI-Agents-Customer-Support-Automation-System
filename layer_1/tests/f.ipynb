```python
unified_ticket = {
  "ticket_id": "TICKET-2001",
  "customer": { "tier": "premium", "lifetime_value": 3000 },
  "message": {
    "normalized": "Bhai, mera refund abhi tak process nahi hua? It has been 10 days! ORD-9988. Very bad service, I am switching to another app.",
    "language": "hinglish"
  },
  "metadata": { "timestamp": "2026-05-07T10:00:00Z" }
}
```

```python
import sys
import os

sys.path.append(os.path.abspath(".."))
```

```python
from app.prompts.supervisor_prompt import SUPERVISOR_PROMPT
from app.services.llm import llm
from app.utils.parser import load_json
from app.models.supervisor_output import SupervisorOutput
```

```python
formatted_prompt = SUPERVISOR_PROMPT.format(
    ticket_id=unified_ticket.get("ticket_id"),
    customer_context=unified_ticket.get("customer"),
    normalized_message=unified_ticket.get("message"),
    metadata=unified_ticket.get("metadata")
)
raw_response = llm.invoke(formatted_prompt)
print("llm output:", raw_response.content)
parsed_data = load_json(raw_response.content)
print("Parsed data:", parsed_data)
supervisor_output = SupervisorOutput(**parsed_data)
print("Supervisor output:", supervisor_output)
```
